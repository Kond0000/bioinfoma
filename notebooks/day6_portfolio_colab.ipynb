# Day6: Portfolio

このノートブックでは、RNA-seq の学習成果を GitHub 上で読みやすい形にまとめる考え方を整理します。

このトピックが役立つ場面:
自分の解析を指導教員、共同研究者、採用担当者に見せるときに、何を理解していて何ができるかを短時間で伝えたい場面です。

このトピックで解く課題:
このRNA-seqケーススタディで何が分かり、なぜその結論に至ったのかを README と図だけで伝わる形にまとめます。

## キーワード辞典（この章）

| キーワード | まずこう理解する | この章での使いどころ | よくある誤解 |
|---|---|---|---|
| research question | この解析で答える問い | README冒頭の芯 | ツール説明を目的と取り違える |
| methods | 何をどう解析したか | 再現性の説明 | ライブラリ名だけ書けば十分と思う |
| results | 図表と主要観察 | 主張の根拠 | 図だけ置いて本文説明を省く |
| limitation | 言える範囲の境界 | 過剰主張の防止 | 書くと弱く見えるから省く |
| reproducibility | 他者が再実行できること | 研究職スキルの核 | ノートブックが1つあれば十分と思う |



## この章での判断軸

1. 5分で第三者が目的・方法・結論を追える構成か。
2. 図に対応する本文解釈があるか。
3. 限界と次の検証を明示しているか。


## ポートフォリオで大事なこと

ポートフォリオは、コードを置くだけでは伝わりません。見る人は、短時間で「何を解析したのか」「どんな流れで進めたのか」「どこまで理解しているのか」を知りたいからです。

## README に入れたい要素

- 解析の目的
- 使ったデータの概要
- 解析の流れ
- 主要な図
- 結果の短い要約
- 今後の拡張点

## 良い見せ方の例

1. 最初に一文でテーマを書く
2. 解析ステップを番号付きで並べる
3. PCA や volcano plot のような代表図を見せる
4. その図から何を読み取ったかを二、三文で添える

## このリポジトリでの目標

この教材では、各トピックを Colab ノートブックとして独立に見られるようにしつつ、トップ README から全体像も追える構成を目指しています。学習ノートではなく、理解の流れが見える資料として仕上げるのが狙いです。

## ここまでの要点

- ポートフォリオはコードだけでなく説明の設計が大事
- 図と文章が対応していると理解度が伝わりやすい
- 読み手は短時間で全体像をつかみたい
- 良い教材は、そのまま良いポートフォリオにもなりやすい
