# Day6: Portfolio

このノートブックでは、RNA-seq の学習成果を GitHub 上で読みやすい形にまとめる考え方を整理します。

このトピックが役立つ場面:
自分の解析を指導教員、共同研究者、採用担当者に見せるときに、何を理解していて何ができるかを短時間で伝えたい場面です。

このトピックで解く課題:
このRNA-seq解析で何が分かり、なぜその結論に至ったのかを README と図だけで伝わる形にまとめます。

## ポートフォリオで大事なこと

ポートフォリオは、コードを置くだけでは伝わりません。見る人は、短時間で「何を解析したのか」「どんな流れで進めたのか」「どこまで理解しているのか」を知りたいからです。

## README に入れたい要素

- 解析の目的
- 使ったデータの概要
- 解析の流れ
- 主要な図
- 結果の短い要約
- 今後の拡張点

## 良い見せ方の例

1. 最初に一文でテーマを書く
2. 解析ステップを番号付きで並べる
3. PCA や volcano plot のような代表図を見せる
4. その図から何を読み取ったかを二、三文で添える

## このリポジトリでの目標

この教材では、各トピックを Colab ノートブックとして独立に見られるようにしつつ、トップ README から全体像も追える構成を目指しています。学習ノートではなく、理解の流れが見える資料として仕上げるのが狙いです。

## ここまでの要点

- ポートフォリオはコードだけでなく説明の設計が大事
- 図と文章が対応していると理解度が伝わりやすい
- 読み手は短時間で全体像をつかみたい
- 良い教材は、そのまま良いポートフォリオにもなりやすい
