## なぜ機能解釈が必要か

差次的発現解析の結果は、多くの場合「遺伝子名の一覧」です。しかし研究では、それだけでは不十分です。重要なのは、その遺伝子群がどんな生物学的変化を示しているかを説明することです。

```python
import pandas as pd

up_genes = ["IL6", "MYC", "TP53"]

annotation = pd.DataFrame({
    "gene": ["IL6", "MYC", "TP53", "GAPDH", "ACTB"],
    "example_function": [
        "inflammatory signaling",
        "cell growth and proliferation",
        "stress response and cell cycle control",
        "glycolysis",
        "cytoskeleton"
    ]
}).set_index("gene")

annotation.loc[up_genes]
```

## どう考えるか

例えば上昇遺伝子の中に `IL6` があれば、炎症応答やサイトカイン関連の変化を連想できます。`MYC` があれば細胞増殖、`TP53` があればストレス応答や細胞周期制御が関係しているかもしれません。

実際の解析では、こうした個別知識に加えて、GO enrichment や pathway 解析を使って「遺伝子群としてどんな機能が偏っているか」を見ます。

## 良い要約の形

良い要約は、遺伝子名の羅列ではなく、条件変化と生物学的意味がつながっています。

例:

treated 条件では `IL6`、`MYC`、`TP53` などが上昇しており、炎症応答と細胞ストレス応答の活性化が示唆される。


## ここまでの要点

- 遺伝子名の一覧だけでは研究の解釈として弱い
- 機能解釈では、遺伝子群から共通する生物学的テーマを探す
- GOや経路解析はその補助になる
