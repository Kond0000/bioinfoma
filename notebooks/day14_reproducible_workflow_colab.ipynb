## この章で解く現実の課題

半年後の自分や共同研究者が、同じ入力から同じ図と表を再現できる解析になっているかを点検します。

研究解析は「一度動いた」で終わりではありません。入力データ、コード、環境、出力、解釈が対応づいていないと、結果の検証も共有もできません。


## 再現性で必要なもの

- 入力データの出典: GEO/SRA ID、取得日、除外理由。
- 実行環境: Python/Rのバージョン、主要ライブラリ。
- 実行手順: どのnotebookをどの順番で実行するか。
- 出力: 図、表、ログ。
- 乱数seed: PCAの初期化やシミュレーションで結果を固定する。
- README: 目的、方法、結果、限界を第三者が読める形にする。

## GitHub READMEに最低限入れる内容

READMEは飾りではなく、解析の入口です。以下がないと、第三者は結果を評価できません。

- Research question: 何を明らかにしたいのか。
- Data: どのデータを使ったのか。
- Methods: count matrix、正規化、PCA、DEG、enrichmentなど何をしたのか。
- Results: どの図や表を見れば結論が分かるのか。
- Interpretation: 生物学的にどう読めるのか。
- Limitations: サンプル数、batch、公共データの制約など。


```python
readme_outline = """
# RNA-seq inflammation case study

## Research question
炎症刺激を受けた細胞で、どの遺伝子群と生物学的経路が変化したのか。

## Data
- Source: GEO/SRA accession
- Samples: control n=3, treated n=3
- Inclusion criteria: same cell type, same time point, same library layout

## Methods
- QC of count matrix
- Normalization and log transformation
- PCA and heatmap
- Differential expression
- Pathway enrichment

## Key results
- Treated samples separated from controls in PCA
- Inflammatory cytokine genes increased
- Inflammatory response pathways were enriched

## Limitations
- Small sample size
- Public metadata may be incomplete
- Experimental validation is not included
"""
print(readme_outline)
```

## エラーが出た場合

- 同じ結果にならない: 乱数seed、ライブラリバージョン、入力データが同じか確認します。
- Colabリンクが開けない: GitHubにpush済みか、URLの `ユーザー名/リポジトリ名/blob/main/...` が合っているか確認します。
